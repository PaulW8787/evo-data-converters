```python
import typing


import glob
from pprint import pp
from evo.data_converters.gef.converter import parse_gef_files

# Data files in data/input have been synthentically generated from a python script.
gef_files = glob.glob("data/input/*.gef")
parsed_cpt_files = parse_gef_files(gef_files)

exclude_keys = [
    "alias",
    "bro_id",
    "column_void_mapping",
    "data",
    "delivered_location",
    "delivered_vertical_position_datum",
]

for cpt_id, cpt_data in parsed_cpt_files.items():
    pp(vars(cpt_data))
    filtered_hash: dict[str, typing.Any] = {k: v for k, v in vars(cpt_data).items() if k not in exclude_keys and v}
    pp(filtered_hash)
```
